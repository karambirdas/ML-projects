```python
import pandas as pd
import sklearn
import numpy as np
```

```python
data=pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
data.head(5)
```

```python
for col in data.columns:
    if len(data[col].unique()) <10 and col!='Churn' :
        data=pd.get_dummies(data, columns=[col], dtype=int)
data=data[data["TotalCharges"]!=" "]
data["TotalCharges"]=pd.to_numeric(data["TotalCharges"],errors='coerce')
data.replace({"Churn":{"Yes":1,"No":0}},inplace=True)
```

```python
X = data.drop(columns=["customerID","Churn"])
y = data["Churn"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.05, random_state=42)
```

```python
#dataframe to numpy array
X_np = X_train.to_numpy()
Data_y = y_train.to_numpy()
#transpose of X
Data_X = X_np.T
#stacking a row of ones
row_1=np.ones((1,Data_X.shape[1]))
Data_X=np.vstack((row_1,Data_X))
```

```python
Data_X
```

```python
Data_X.shape
```

```python
#sigmoid function
def sigmoid(z):
    return 1.0 / (1.0 + z)
```

```python
#creating the thetas
theta=np.zeros((Data_X.shape[0],1))
```

```python
#h_theta function
def h_theta(theta,Data_X):
    z=np.dot(theta.T,Data_X)
    return sigmoid(z)
h_theta(theta,Data_X)
```

```python
#gradient function
def gradient(theta,Data_X,Data_y):
    grad=(1/Data_X.shape[1])*np.dot((Data_y - h_theta(theta,Data_X)), Data_X.T)
    return grad
```

```python
gradient(theta,Data_X,Data_y)
```

```python
#gradient descent function
def gradient_descent(theta,Data_X,Data_y,learning_rate,num_iterations):
    for i in range(num_iterations):
        grad=gradient(theta,Data_X,Data_y)
        theta=theta + learning_rate*grad.T
    return theta
```

```python
theta_opt=gradient_descent(theta,Data_X,Data_y,0.01,1000)
```

```python
y_test.shape
```

```python
X_test.shape
```

```python
#dataframe to numpy array
X_test_np = X_test.to_numpy()
Y_test_np = y_test.to_numpy()
#transpose of X
X_test_1 = X_test_np.T
row_1=np.ones((1,X_test_1.shape[1]))
X_test_1=np.vstack((row_1,X_test_1))
```

```python
X_test_1
```

```python
X_test_1.T[1]
```

```python
y_pred=h_theta(theta_opt,X_test_1)
```

```python
y_pred
```

```python
y_p=np.where(y_pred>=0.5,1,0)
```

```python
y_p
```

```python
#compare y_pred with y_test
matches= y_p == Y_test_np
```

```python
#true positive estimation
TP_plus_TN=np.sum(matches)
```

```python
Accuracy=TP_plus_TN/Y_test_np.shape[0]
Accuracy   
```

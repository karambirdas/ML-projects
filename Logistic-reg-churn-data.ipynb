```python
import pandas as pd
import sklearn
```

```python
data=pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
#Add the dummy columns where the number of unique values is less than 10
for col in data.columns:
    if len(data[col].unique()) <10 and col!='Churn' :
        data=pd.get_dummies(data, columns=[col], dtype=int)
```

```python
#Convert the TotalCharges column to numeric
data=data[data["TotalCharges"]!=" "]
data["TotalCharges"]=pd.to_numeric(data["TotalCharges"],errors='coerce')
```

```python
data.head()
```

```python
#Dropping the unnecessary columns
data.replace({"Churn":{"Yes":1,"No":0}},inplace=True)
X = data.drop(columns=["customerID","Churn"])
y = data["Churn"]
```

```python
#Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.05, random_state=42)
```

```python
#Training the Logistic Regression model
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression(max_iter=10000)

model.fit(X_train, y_train)
```

```python
#Evaluating the model
y_pred = model.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
